```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
def congruencial_multiplicativo(a, m, x0, n):
  numeros_aleatorios = []
  #numeros_redondeados = []
  x = x0

  for _ in range(n):
    x = (a * x) % m
    numeros_aleatorios.append(x / m)  # Normalizamos para obtener valores entre 0 y 1

    #numeros_redondeados = [round(numero, 4) for numero in numeros_aleatorios]

  return numeros_aleatorios

# parametros 
a = 5
m = 11
x0 = 7
n = 10

resultados = congruencial_multiplicativo(a, m, x0, n)
print(resultados)
print("Cantidad de numeros aleatorios: ",len(resultados))
type(resultados)
```

```python
plt.hist(resultados, bins=10
         , color='blue', edgecolor='black')
plt.title("Histograma de números aleatorios")
plt.xlabel("Intervalos")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Prueba de Kolmogorov-Smirnov
D, p_valor = stats.kstest(resultados, 'uniform')
print("Estadístico de Kolmogorov-Smirnov:", D)
print("p-valor:", p_valor)
```

```python
if p_valor < 0.05:
    print("Los datos no parecen provenir de una distribución uniforme.")
else:
    print("No se puede rechazar la hipótesis nula. Los datos podrían ser uniformes.")
```
